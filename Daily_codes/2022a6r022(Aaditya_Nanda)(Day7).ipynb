```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
ship = pd.read_csv("titanic.csv")
ship
```

# Graph comparsion between Male , Female and child survival 

```python
servival = ship.groupby(["who","alive"]).size().unstack(fill_value=0)
servival
```

```python
servival.plot(kind='bar', color=['Red','Green'])
plt.grid()
plt.show()
```

# Graph between People and Class

```python
sales = ship.groupby(["class","who"]).size().unstack(fill_value=0)
sales
```

```python
sales.plot(kind='bar', color=['Red','Green','blue'])
plt.grid()
plt.show()
```

# Graph between Alone and Alive

```python
life = ship.groupby(["alone","alive","who"]).size().unstack(fill_value=0)
life 
```

```python
life.plot(kind='bar', color=['Red','Green','blue'])
plt.grid()
plt.show()
```
